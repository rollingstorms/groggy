```python
from groggy import Graph
from groggy.algorithms import community

sub = g.view()
pipe = community.connected_components(output_attr="component")
# pipe = community.lpa()

# Persisting results
_, persist_profile = sub.apply(pipe, return_profile=True, persist=True)

# View-only (no set_node_attrs)
_, view_profile = sub.apply(pipe, return_profile=True, persist=False)

print("Persist timers:", persist_profile["timers"])
print("View timers:", view_profile["timers"])
print("Persist run_time:", persist_profile["run_time"])
print("View run_time:", view_profile["run_time"])
```
